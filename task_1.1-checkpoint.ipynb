```python

'''     Задача 2. Раздел 1.2. Условные конструкции. Операторы сравнения.

                                Второй вариант.

    Пример для военкомата.

Проверьте пригодность к службе в определённых войсках для призывника, в зависимости от:
1. Роста
2. Возраста
3. Количества детей
4. Учится ли он сейчас
'''
print('\n    Задача 2. Раздел 1.2\n')
height = int(input(' Введите рост призывника: '))
if height < 160:
    print('\n К службе не годен.')
elif 18 > int(input(' Введите возраст призывника: ')) > 27:
    print('\n К службе не годен.')
elif int(input(' Укажите количество детей: ')) >= 2:
    print('\n К службе не годен.')
elif input(' Учится ли он сейчас (y/n): ') == 'y':
    print('\n К службе не годен.')
else:
    if 160 <= height <= 169:
        print('\n Годен. Танковые войска.')
    elif 170 <= height <= 179:
        print('\n Годен. Морской флот.')
    elif 180 <= height <= 200:
        print('\n Годен. Десантные войска.')
    elif height > 200:
        print('\n Годен. Другие войска.')

print('\n  -- Конец --  ')  #               - Для блокнота
#input('\n  -- Конец --  ')  # Типа  "Пауза" - Для среды
```

```python

'''     Задача 2. Раздел 1.2. Условные конструкции. Операторы сравнения.

                                Третий вариант.

    Пример для военкомата.

Проверьте пригодность к службе в определённых войсках для призывника, в зависимости от:
1. Роста
2. Возраста
3. Количества детей
4. Учится ли он сейчас
'''
print('\n    Задача 2. Раздел 1.2\n')
height = int(input(' Введите рост призывника: '))
old = int(input(' Введите возраст призывника: '))
children = int(input(' Укажите количество детей: '))
student = input(' Учится ли он сейчас (y/n): ')
if height >= 160 and 18 <= old <= 27 and children < 2 and student == 'n':
    if 160 <= height <= 169:
        print('\n Годен. Танковые войска.')
    elif 170 <= height <= 179:
        print('\n Годен. Морской флот.')
    elif 180 <= height <= 200:
        print('\n Годен. Десантные войска.')
    elif height > 200:
        print('\n Годен. Другие войска.')
else:
    print('\n К службе не годен.')

print('\n  -- Конец --  ')  #               - Для блокнота
#input('\n  -- Конец --  ')  # Типа  "Пауза" - Для среды
```
